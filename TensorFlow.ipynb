```python
x = tf.placeholder(tf.int32)
m = tf.Variable(30)
c = tf.Variable(40)
var = tf.compat.v1.global_variables_initializer()
with tf.compat.v1.Session() as sess:
  y=m*x+c
  sess.run(var)
  sess.run(m)
  sess.run(c)
  print(sess.run(y,{x:10}))
  print(sess.run(y,{x:[1,2,3,4,5]}))
```
